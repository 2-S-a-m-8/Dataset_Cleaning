```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Samanyu K\Downloads\Customer Call List.csv")
df
```

```python
df=df.drop_duplicates()
cols=['Not_Useful_Column',	'Unnamed: 8',	'Unnamed: 9']
df=df.drop(columns=cols)
df
```

```python
df['Last_Name']=df['Last_Name'].str.replace('[/._]','',regex=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace(r'\D','',regex=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].fillna("")
df['Phone_Number']=df['Phone_Number'].apply(lambda x:'+' + x[0:2]+" "+x[2:])
df['Phone_Number']=df['Phone_Number'].apply(lambda x:" " if x.strip()=="+" else x)
df
```

```python
df[['Street', 'State', 'Pincode']] =df['Address'].str.split(',', n=2, expand=True)
df=df.drop(columns=['Address'])
df
```

```python
df['Paying Customer']=df['Paying Customer'].apply(lambda x:'Y' if x=='Yes' else  'N' if   x=='No' else x)
df['Do_Not_Contact']=df['Do_Not_Contact'].apply(lambda x:'Y' if x=='Yes' else  'N' if   x=='No' else x)
df
```

```python
df=df.fillna('')
df['Paying Customer']=df['Paying Customer'].str.replace('N/a','')
df['Street']=df['Street'].str.replace('N/a','')
df
```

```python
df=df[df['Phone_Number']!=" "]
df.reset_index(drop=True,inplace=True)
df
```

```python
df=df.rename(columns={"Phone_Number":"Contact Number","Do_Not_Contact":"Contact Status"})
df
```

```python
df['Contact Status']=df['Contact Status'].apply(lambda x:"Y" if x=="" else x)
df
```

```python
df=df.sort_values(by=['Paying Customer'],ascending=True)
df.reset_index(drop=True,inplace=True)
df
```
