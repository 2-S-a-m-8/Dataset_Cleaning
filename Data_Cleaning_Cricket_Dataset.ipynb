```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Samanyu K\Downloads\Cricket_dataset.csv")
df
```

```python
df=df.rename(columns={"Inns":"Innings",'Mat':'Matches','NO':'Not_Out','HS':'Highest_Innings_Score','BF':'Balls_Faced','SR':'Batting_Strike_Rate','Ave':'Average'})
df
```

```python
pd.set_option('Display.max_rows',None)
df.shape
```

```python
df.isnull().any()
```

```python
df=df.drop_duplicates()
df.shape
```

```python
df[['Start_Year','End_Year']]=df['Span'].str.split('-',n=1,expand=True)
df
```

```python
d=df['Start_Year'].dtype
d1=df['End_Year'].dtype
print(d,d1)
#Converting the above datatypes to string for calculation of Career Length
df['Start_Year']=df['Start_Year'].astype(int)
df['End_Year']=df['End_Year'].astype(int)
d2=df['Start_Year'].dtype
d3=df['End_Year'].dtype
print(d2,d3)
```

```python
df['Career_Length']=df['End_Year']-df['Start_Year']
df=df.drop(columns=['Span'])
df
```

```python
df[['Player','Country']]=df['Player'].str.split('(',n=1,expand=True)
df['Country']=df['Country'].str.replace(')','')
df
```

```python
df.dtypes
```

```python
print(df['Highest_Innings_Score'])
#Fomrmatting 'Highest_Innings_Score'
df['Highest_Innings_Score']=df['Highest_Innings_Score'].replace(r'\D','',regex=True)
df['Highest_Innings_Score']=df['Highest_Innings_Score'].astype(int)
print(df['Highest_Innings_Score'])
print(df['Highest_Innings_Score'].isnull().any())
df.dtypes
```

```python
#formatting balls faced
print(df['Balls_Faced'])
df['Balls_Faced']=df['Balls_Faced'].replace(r'\D','',regex=True)
print(df['Balls_Faced'])
df['Balls_Faced']=df['Balls_Faced'].apply(lambda x: 0 if x=="" else x)
print(df['Balls_Faced'])
df['Balls_Faced']=df['Balls_Faced'].astype(int)
print(df['Balls_Faced'].dtype)
```

```python
df.dtypes
#formatting 4s and 6s and changing datatypes
df['4s']=df['4s'].str.replace(r'\D','',regex=True)
print(df['4s'])
df['6s']=df['6s'].str.replace(r'\D','',regex=True)
print(df['6s'])
df['4s']=df['4s'].astype(int)
df['6s']=df['6s'].astype(int)
```

```python
df['100']=df['100'].astype(int)
df['50']=df['50'].astype(int)
df['0']=df['0'].astype(int)
df.dtypes
```

```python
# Average Career length of the cricketers in the given dataframe
avg=df['Career_Length'].mean()
print(avg)
```

```python
# Find number of cricketers who played before 1960
c=df.loc[df['Start_Year']<1960]['Player'].count()
print(c)
```

```python
# Max. Highest Innings score by country
df1=df.groupby(df['Country'])['Highest_Innings_Score'].max()
print(df1)
#Sorting the dataframe based on Highest innings score from max to min
df1=df1.sort_values(ascending=False)
print(df1)
```

```python
#Finding average of 100's, 50's and ducks (0's) Average by country
df2=df.groupby(df['Country'])[['100','50','0']].aggregate('mean')
print(df2)
```

```python
df=df.sort_values(by=['Career_Length','Runs'],ascending=False).reset_index(drop=True)
df
```
